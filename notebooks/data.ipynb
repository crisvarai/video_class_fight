```python
import pandas as pd
import os
```

```python
dir_path = '../Peliculas_New/'
```

```python
def get_dataframe(path):
    videos = []
    labels = []
    for folder in os.listdir(path):
        # print(folder)
        fd = dir_path + folder + '/'
        # print(fd)
        for video in os.listdir(fd):
            vd = os.path.join(fd, video)
            # print(vd)
            i = 1 if folder == 'fights' else 0
            videos.append(vd)
            labels.append(i)
    data_dict = {
        'videos': videos,
        'labels': labels 
    }
    dataframe = pd.DataFrame(data=data_dict)
    return dataframe
```

```python
df = get_dataframe(dir_path)
df.head(), df.tail()
```

```python
vid_length = [1] * len(df)
len(vid_length)
```

```python
df['video_length'] = vid_length
df.head(), df.tail()
```
